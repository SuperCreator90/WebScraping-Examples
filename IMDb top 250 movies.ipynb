```python
from  bs4 import BeautifulSoup as bs 
# small caps --> lib name, capital --> Class name, as --> alias --> hum payar se bulate hai
import pandas as pd
import requests 
import re
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.76 Safari/537.36'}
```

```python
url = 'https://www.imdb.com/chart/top/?ref_=nv_mv_250'
```

```python
webpage = requests.get(url, headers=headers).text
soup = bs(webpage, 'lxml')
```

```python
res=soup.find('tbody',class_='lister-list') # find all takes two input parameters --> html tags, class, id, href
```

```python
res
```

```python
c = 1
for i,j in zip(res.find_all('td',class_='titleColumn'),res.find_all('td',class_='imdbRating')):
    print(c," ".join(i.text.strip().split()).split('(')[0].split('.')[1].strip()," ".join(i.text.strip().split()).split('(')[1].strip(')'), j.text.strip())
    c += 1
```

```python
m_no = []
m_nm = []
m_rd = []
m_rt = []
c = 1

for i,j in zip(res.find_all('td',class_='titleColumn'),res.find_all('td',class_='imdbRating')):
    m_no.append(c)
    m_nm.append(" ".join(i.text.strip().split()).split('(')[0].split('.')[1].strip())
    m_rd.append(" ".join(i.text.strip().split()).split('(')[1].strip(')'))
    m_rt.append(j.text.strip())
    c += 1
```

```python
len(m_no)
```

```python
df = pd.DataFrame()
```

```python
df['movie_no'] = m_no
df['movie_name'] = m_nm
df['release_date'] = m_rd
df['rating'] = m_rt
```

```python
df.to_csv('imdb_top_250_movies.csv',index=False)
```
